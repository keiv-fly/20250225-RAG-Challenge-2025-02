```python
import openai
import tqdm
import pandas as pd
```

```python
with open("../tokens/openai_token.txt") as f:
    openai_token = f.read()
client = openai.OpenAI(
    # This is the default and can be omitted
    api_key=openai_token,
)
```

```python
response = client.files.list(purpose="assistants")
files = list(response.data)
```

```python
df = pd.DataFrame([file.model_dump() for file in files])
df = df.sort_values("bytes", ascending=False)
df
```

```python
for i, file_i in enumerate(tqdm.notebook.tqdm(df.index.to_list())):
    file = files[file_i]
    if i % 10 == 0:
        print(f"{i}: {file}")
    try:
        client.files.delete(file.id)
    except openai.NotFoundError as e:
        print(f"Not found: {file}")
```
