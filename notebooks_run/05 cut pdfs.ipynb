```python
import pandas as pd

from pypdf import PdfReader, PdfWriter
import picologging as logging

import pathlib
import datetime
```

```python
now = datetime.datetime.now()
now_str = now.strftime("%Y%m%d_%H%M%S")
pathlib.Path(f"../notebooks_logging/05_cut_pdfs").mkdir(parents=True, exist_ok=True)
# pathlib.Path(f"../notebooks_logging/05_cut_pdfs/{now_str}").mkdir(parents=True, exist_ok=True)

file_handler = logging.FileHandler(f"../notebooks_logging/05_cut_pdfs/{now_str}.log")
stream_handler = logging.StreamHandler()
# stdout_formatter = logging.Formatter(fmt="%(message)s") 
# stream_handler.setFormatter(stdout_formatter)
logging.basicConfig(
    encoding='utf-8', 
    format='%(asctime)s | %(levelname)s | %(message)s', 
    level=logging.INFO, 
    datefmt='%Y-%m-%dT%H:%M:%S',
    handlers=[
        file_handler,
        stream_handler,
    ],
    force=True
)
logger = logging.getLogger()
logger.setLevel(logging.INFO)
li = logger.info
li("start")
```

```python
df_subset = pd.read_parquet(f"../data_temp/04 subset.parquet")
df_subset.columns
# df_subset[["sha1", "is_in_questions"]]
```

```python
df_subset_filtered = df_subset[df_subset["is_in_questions"]==True]
df_subset_filtered
```

```python
li(df_subset_filtered)
```

```python
pdf_name_i = df_subset_filtered["sha1"].iloc[0] + ".pdf"
pdf_name_i
```

```python
li("Cutting pdfs:")

total_pages = 0

for id_row, row in df_subset_filtered.iterrows():
    pdf_name_i = row["sha1"]
    pdf_name_i_w_pdf = pdf_name_i + ".pdf"
    li(f"pdf_name_i={pdf_name_i_w_pdf}")

    reader = PdfReader(f"../data_in/temp/pdfs/pdfs/{pdf_name_i_w_pdf}")

    pathlib.Path(f"../data_temp/05_cut_pdfs/{pdf_name_i}").mkdir(parents=True, exist_ok=True)
    for i, page in enumerate(reader.pages):
        writer = PdfWriter()
        writer.add_page(page)
        with open(f"../data_temp/05_cut_pdfs/{pdf_name_i}/{pdf_name_i}_{i}.pdf", 'wb') as out:
            writer.write(out)
        li(f"page={i}, path=../data_temp/05_cut_pdfs/{pdf_name_i}/{pdf_name_i}_{i}.pdf")
        total_pages += 1

li(f"total_pages={total_pages}")
```
